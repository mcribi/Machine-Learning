Vamos a explicar más en detalle esta interacción y explicar que es exactamente el **ciclo de interacción entre el agente y el entorno**.

Nuestra meta es crear algoritmos en el agente para que tome decisiones (elija acciones) con la información que tiene en cada momento de instante $t$. Por otro lado, a la derecha tenemos el mundo que equivale al entorno que es todo lo que no es el agente, pero que interacciona con el agente. Por tanto, estamos ante una comunicación bidireccional porque tanto el agente se comunica con el entorno como el entorno se comunica con el agente. Todas esas comunicaciones de las que hablamos son tres señales: acciones, recompesas y observaciones (como se puede observar en la ilustración).

Un ciclo de interacción es el siguiente: el agente solo puede influir en el entorno por medio de las acciones $A_t$ elegidas en cada instante $t$, y el entorno responde a estas acciones proporcionando una observación $O_{t+1}$ y una señal de recompensa $R_{t+1}$. La observación se puede entender como una fotografía del entorno en ese instante. Esquemáticamente:

En cada instante *t* el agente:
- Recibe la observación $O_t$.
- Recibe la recompensa $R_t$.
- Ejecuta la acción $A_t$.

Por otra parte, el entorno:
- Recibe la acción $A_t$.
- Emite la observación $O_{t+1}$.
- Emite la recompensa $R_{t+1}$.

La actualización del instante de tiempo se produce en el entorno porque cada vez que genera una nueva observación y emite una nueva recompensa.  

Esta interacción con estos pasos se va a repetir donde el agente mediante el procedimiento de prueba y error, irá aprendiendo.


### Un estado y el espacio de estados.

Antes de definir un estado y el espacio de estados vamos a definir lo que es la historia. La historia es la secuencia de todas las acciones, observaciones y recompensas desde el inicio hasta el instante $t$:

$$
H_t = \{ O_0, A_0, R_1, O_1, A_1, R_2, O_2, A_2, \dots, A_{t-1}, R_t, O_t \}
$$

Es decir, la historia esta formada por todas las variables que son observadas a lo largo del tiempo por el agente y por el entorno. Por tanto, el agente elijirá la acción basándose es la historia. Sin embargo, la historia es una estructura muy grande que se incrementa con el tiempo, lo que lo hace inmanejable. Por tanto, en vez de utilizar esta estructura se utilizan los estados:

El **estado** $S_t$ es la información que se usa para determinar qué sucederá en el próximo instante de tiempo. Formalmente es una función de la historia:

$$
S_t = f(H_t)
$$

Por tanto, el $S_t$ puede ser un resumen de la historia $H_t$ pero que debe incluir toda la información necesaria para determinar qué sucederá a continuación, tanto desde el punto de vista del **agente** (es necesario que el agente pueda tomar decisiones y ejecutar acciones según ese estado), como del **entorno** (el entorno debe generar las próximas observaciones y  las próximas recompensas según el estado).

Así el espacio de estados es el conjunto de todos los posibles estados que el agente puede experimentar en el entorno, es decir, a toda la información proporcionada por el estado de un entorno:

$$
\mathcal{S} = \{ S_t \mid S_t = f(H_t), \ \forall t \}
$$

Puede ser discreto donde el número de estados posibles es finito, como por ejemplo en un tablero de ajedrez. O bien continuo donde los estados se representan en un espacio continuo como por ejemplo, la posición y velocidad de un coche en un entorno físico.

### Una acción y el espacio de acciones.
Hemos hablado que la historia incluye las acciones tomadas, pero no hemos definido exactamente que es una accion. La acción es un movimiento o una decisión que toma el agente para interactuar con el entorno. Cada accion que realiza se mueve el entorno a nuevos estados posteriores.

El espacio de acciones se denota por todas las acciones posibles que el agente puede realizar dentro de un estado.

### Una recompensa
La **señal recompensa ($R_t$)** es un valor escalar que el agente recibe en el
instante $t$ después de cada acción que ejecuta. Tenemos el siguiente teorema (Hipótesis de la recompensa):

> Cualquier objetivo puede describirse mediante la maximización de la  
> esperanza (media estadística) de la recompensa acumulada.  

Dicha recompensa acumulada, también llamada **retorno**,   
puede definirse para el instante *t* como:  

$$
G_t = R_{t+1} + R_{t+2} + R_{t+3} + \dots
$$


Es decir, se define la señal de **retorno** ($G_t$) como el valor acumulado de todas las recompensas futuras.  Remarcar que, como dichas recompensas son aleatorias, el retorno también lo es y por eso el objetivo del agente es maximizar su valor esperado.  

Por otro lado, dado que el número de episodios (cada vez que se recibe una
recompensa después de una acción) puede no ser finito, el sumatorio de  recompensas podría no converger y obtener un valor de retorno infinito.  

En este caso, conviene redefinir el retorno como:  

$$
G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \dots
$$

En la ecuación anterior se introduce un parámetro $(\gamma)$ llamado factor de descuento, cuyo valor está acotado (concretamente $(0 \leq \gamma \leq 1)$) y permite la convergencia del valor del retorno en los problemas donde el número de recompensas futuras es infinito.  


### Una política: políticas deterministas y no deterministas
Una **política** define la forma de actuar del agente de aprendizaje en un momento dado, es decir, es como elige el agente las acciones. A grandes rasgos, una política es una correspondencia entre los estados percibidos del entorno y las acciones que deben emprenderse en esos estados. En algunos casos, la política puede ser una simple función o una tabla de consulta, mientras que en otros puede implicar una gran cantidad de cálculos, como un proceso de búsqueda. La política es el núcleo de un agente de aprendizaje por refuerzo en el sentido de que por sí sola es suficiente para determinar el comportamiento. En general, las políticas pueden ser estocásticas y especificar probabilidades para cada acción.

Viendola como una función ($\pi$), es un mapeo entre los estados $S_t$ y las acciones $A_t$.

Existen dos tipos de políticas:
* Deterministas: cuando es un mapeo directo entre el estado y la acción
$$ a=\pi (s)$$
Como vemos, la acción $a$ está determinada unívocamente por el estado $s$ donde está el agente.
* Estocástica: puede haber más de una acción a escoger para cada estado, por lo que la política asigna una probabilidad a cada opción:
$$ \pi(a | s) = Pr\{A_t = a | S_t = s\} $$

donde $\pi(a | s)$ es la probabilidad de seleccionar la acción $a$ si el agente se encuentra en el estado $s$. Este suele ser el caso habitual en la mayoría de los problemas de la vida real.  

Este tipo de polítca permite que la exploración del entorno sea más estocástica porque al haber varias acciones posibles por estado, si repetimos un estado puede que se tomen dos acciones distintas en cada caso.


### Una experiencia

Las experiencias son denominada como una secuencia de estado estado, acciones y recompensas que se obtienen por muestreo a partir de la interacción real o simulada con un entorno. Es todo lo que necesitan los métodos Monte Carlo, que son formas de resolver el problema del aprendizaje por refuerzo basadas en promediar los rendimientos de las muestras.

Es decir, es cuando hay una interacción completa entre el agente y el entorno en un momento de tiempo determinado: estado actual, acción tomada, recompensa tras la acción y nuevo estado después de la acción.

### Un episodio y una trayectoria.
El agente debe de maximizar la recompensa acumulada a largo plazo. Para ello intenta maximizar el rendimiento ($G_t$) que se puede definir com la suma de recompensas:

$$
G_t = R_{t+1} + R_{t+2} + R_{t+3} + \dots + R_T
$$

donde $T$ es un paso temporal final. Este enfoque tiene sentido en aplicaciones en las que existe una noción natural de paso temporal final, es decir, cuando la interacción agente-entorno se divide de forma natural en subsecuencias, que llamamos **episodios**. Cada episodio termina en un estado especial llamado estado terminal, seguido de un restablecimiento a un estado inicial estándar o a una muestra de una distribución estándar de estados iniciales. Por lo tanto, se puede considerar que todos los episodios terminan en el mismo estado final, con diferentes recompensas para los diferentes resultados. Las tareas con episodios de este tipo se denominan tareas episódicas.

Por otro lado, la trayectoria es la secuencia de estados, acciones y recompensas:

$$S_0, A_0, R_1, S_1, A_1, R_2, S_2, A_2 \dots $$

La diferencia entre episodio y trayectoria es que una trayectoria puede ser infinita, es decir, no tiene porque tener un final pero sí un principio. Además, como hemos definido antes es un registro de la evolución del agente. En un episodio si tiene final porque hemos definido el estado terminal. Cuando llega a éste termina y se reinicia. Se puede representar también como una secuencia de estados, acciones y recompensas, pero la diferencia clave es que está acotado en el tiempo, es decir, $S_T$ es el estado terminal:

$$S_0, A_0, R_1, S_1, A_1, R_2, S_2, A_2,  \dots , S_T$$

###El objetivo del aprendizaje por retorno, el concepto retorno y el factor de descuento

El objetivo de un agente en aprendizaje por retorno, en realidad, el objetivo de cualquier agente no es otro que el de solucionar el problema que se plantea en un entorno. La diferencia entre el agente de aprendizaje por retorno y otro tipo de agente es que su objetivo es maximizar el retorno acumulado a lo largo del tiempo, eso implica considerar el impacto futuro de las acciones actuales. Se podría decir que una de las principales diferencias también es su interacción secuencial con el entorno.

Pero claro, si hablamos de maximización del retorno, primero habría que saber la definición del mismo, por lo que pasamos a contestar una pregunta clave:

> **¿Qué es el retorno?**

Si asumimos *t* como el instante actual y *R* como la recompensa; podemos definir que el retorno como la siguiente *G*:

<center>G<sub>t</sub>=R<sub>t+1</sub>+R<sub>t+2</sub>+R<sub>t+3</sub>+...</center>

Por lo que el retorno de un instante se puede definir como la acumulación de las recompensas de todos los instantes futuros a partir del actual.

Como las recompensas que estamos evaluando son aleatorias, el retorno también lo es; por eso el objetivo del agente es maximizar el mismo.

El principal problema que encaramos entonces sería que el número de episodios no sea finito, lo cual volvería también infinito el valor del retorno. Es por ello que se redefine el retorno tal que:

<center>G<sub>t</sub>=R<sub>t+1</sub>+γR<sub>t+2</sub>+(γ^2)R<sub>t+3</sub>+...</center>

Introducimos el concepto de factor de descuento `γ` acotado entre 0 y 1 ambos incluidos. Esto permite la convergencia del retorno en entornos donde el número de episodios es infinito y a parte influye de una manera muy positiva en el proceso de aprendizaje de modelos que usen el proceso de decisión de Markov.

###Valor de un estado *V(s)*

El valor de un estado, denotado como *V(s)*, es una función que mide el valor esperado del retorno futuro que un agente puede obtener comenzando en el estado s.

Aunque faltaría por definir la política del agente con el que se evalua la función; por convenio se suele utilizar π; por lo tanto se redefiniría la funcion como *Vπ(s)*

Por lo que ya sabemos que esta función es la que se utiliza para evaluar estados y tomar las decisiones a futuro; por lo que solo nos queda observar su definición formal

<center>Vπ(s)=E<sub>π</sub>[G<sub>t</sub> | S<sub>t</sub> = s] = E<sub>π</sub> [R<sub>t+1</sub>+γR<sub>t+2</sub>+(γ^2)R<sub>t+3</sub>+ ... | S<sub>t</sub> = s]</center>

Donde:

  * **E<sub>π</sub>** es la esperanza matemática de la política π

  * **G<sub>t</sub>** es el retorno desde el tiempo *t*

  * **γ** es el factor de descuento definido con anterioridad

  * **R<sub>t+k</sub>** son las recompensas en los episodios siguientes

### Par estado-acción *Q(s,a)*

Una vez vista la función de valor de un estado tenemos que ser caoaces de evaluar también la dupla valor-acción; es por ello que se define la siguiente función en la que se evalua dicha pareja representada como *q<sub>π</sub>(s,a)*:

<center>Q<sub>π</sub>(s,a) = E<sub>π</sub>[G<sub>t</sub> | S<sub>t</sub> = s, A<sub>t</sub> = a] = E<sub>π</sub> [R<sub>t+1</sub> + γR<sub>t+2</sub> + (γ<sup>2</sup>)R<sub>t+3</sub> + ... | S<sub>t</sub> = s, A<sub>t</sub> = a]</center>

Con esta función a parte de todo lo anterior se suma el calculo de si el agente en estado *s* toma la decisión *a* siguiendo su política *π*, la cual continuará siguiendo después de llevar a cabo la acción. Al ser una función tan completa, muchos de los algoritmos principales de aprendizaje por refuerzo tienen como base una política que consiste en maximizar el valor de esta función.

**Política aleatoria**

Usando de nuevo la biblioteca numpy como en el resto de prácticas implementamos la política aleatoria:

````python
import numpy as np
import random

def random_policy(env, episodes=100, render=False):
    env = TimeLimit(env, max_episode_steps=500)

    rewards = []
    for ep in range(episodes):
        obs, _ = env.reset()
        total_reward = 0
        done = False

        while not done:
            action = env.action_space.sample()
            obs, reward, terminated, truncated, _ = env.step(action)
            total_reward += reward
            done = terminated or truncated

            if render and ep == 0:
                print(env.render())

        rewards.append(total_reward)
    return rewards
````

**Política óptima**

Dado que el entorno es una cuadrícula con un acantilado peligroso en la fila inferior, esta política evita esa zona moviéndose por la fila segura justo encima (fila 2) hacia la derecha. Al llegar a la última columna, el agente desciende a la fila final para alcanzar el objetivo. Esta estrategia evita por completo el acantilado y sigue el camino más corto posible, lo que permite obtener la recompensa óptima de -13. Aunque está codificada manualmente y no es aprendida, representa el comportamiento ideal en este entorno.

````python
from gymnasium.wrappers import TimeLimit

def optimal_policy(env, episodes=100, render=False):
    env = TimeLimit(env, max_episode_steps=500)

    rewards = []
    for ep in range(episodes):
        obs, _ = env.reset()
        total_reward = 0
        done = False

        while not done:
            row = obs // 12
            col = obs % 12

            # Política: ir por fila 2 y bajar al final
            if row == 3:
                action = 0  # subir
            elif row == 2 and col < 11:
                action = 1  # derecha
            elif row == 2 and col == 11:
                action = 2  # abajo (llegar a la meta)
            else:
                action = 1  # fallback: derecha

            obs, reward, terminated, truncated, _ = env.step(action)
            total_reward += reward
            done = terminated or truncated

            if render and ep == 0:
                print(env.render())

        rewards.append(total_reward)
    return rewards
````

**Ejecución e impresión de resultados**

````python
random_rewards = random_policy(env, 100)
optimal_rewards = optimal_policy(env, 100)

print("Recompensa promedio (política aleatoria):", np.mean(random_rewards))
print("Recompensa promedio (política 'óptima'):", np.mean(optimal_rewards))
````

La optimal policy nos revela que hacen falta exactamente 13 pasos para llegar a la meta (-1 por cada paso dado)

####Algoritmos clásicos de aprendizaje por refuerzo

El Aprendizaje por Refuerzo (Reinforcement Learning, RL) se enfocaba en algoritmos tabulares diseñados para resolver problemas con espacios de estados y acciones relativamente pequeños; o al menos eso hacía antes de que el Deep Learning se volviese tremendamente popular, de ahí que se les llame "clásicos".

Estos métodos se basaban en la teoría de Procesos de Decisión de Markov (MDP), en los que un agente aprende a tomar decisiones mediante la interacción con un entorno para maximizar una recompensa acumulada a lo largo del tiempo.

**Métodos clásicos más representativos:**

  * **Value Iteration:** Calcula los valores de los estados de forma iterativa hasta la convergencia, y luego deriva una política óptima.

  * **Policy Iteration:** Alterna entre evaluar una política y mejorarla hasta encontrar la óptima.

  * **Q-Learning:** Aprende valores de acción (estado-acción) mediante actualizaciones basadas en la experiencia, sin necesidad de conocer el modelo del entorno.

  * **SARSA:** Similar a Q-Learning, pero actualiza los valores usando la acción realmente tomada por la política del agente.

  * **Métodos Monte Carlo:** Aprenden a partir de episodios completos, actualizando estimaciones basadas en la recompensa total obtenida.

Estos serían los algoritmos más importantes, los cuales permitieron desarrollar políticas efectivas sin el uso de redes neuronales ni técnicas de representación complejas, y siguen siendo fundamentales para comprender el núcleo del aprendizaje por refuerzo.



####Selección y descripción de un algoritmo: Q-Learning

**Descripción**

Q-Learning es un algoritmo de aprendizaje por refuerzo off-policy que busca aprender la función óptima de valor de acción Q ∗ (s,a), la cual indica el valor esperado de tomar una acción *a* en un estado *s* y seguir la mejor política posible a partir de ahí.

El algoritmo no requiere conocer el modelo del entorno (es decir, las probabilidades de transición y las recompensas), lo que lo hace muy flexible.



**Función de actualización**

La función de actualización de Q-Learning es la siguiente:

  <center>Q(s<sub>t</sub>,a<sub>t</sub>)←Q(s<sub>t</sub>,a<sub>t</sub>)+α[r<sub>t+1</sub>+γ*max<sub>a′</sub>Q(s<sub>t+1</sub>​,a′)−Q(s<sub>t</sub>,a<sub>t​</sub>)]</center>

Donde:

**Q(s<sub>t</sub>,a<sub>t</sub>)**: Valor actual estimado de tomar acción a<sub>t</sub> en estado s<sub>t</sub>

**α**: Tasa de aprendizaje (0<α≤1), controla cuánto se actualiza el valor anterior

**r<sub>t+1</sub>**: Recompensa obtenida al tomar la acción

**γ**: Factor de descuento (0≤γ≤1), valora la importancia de las recompensas futuras

**max<sub>a′</sub>Q(s<sub>t+1</sub>​,a′)**: Valor máximo estimado de la mejor acción posible en el siguiente estado

Esta fórmula ajusta el valor estimado de una acción en función de la recompensa inmediata y la mejor expectativa futura.

**Pasos del algoritmo**

1.   Inicialización:

      Crear una tabla Q(s,a) con valores iniciales arbitrarios (usualmente ceros).

2.   Para cada episodio:
      * Inicializar el estado s.

      * Repetir hasta que el estado sea terminal:
          * Seleccionar una acción a usando una política de exploración.
          * Ejecutar la acción a y observar la recompensa r y el nuevo estado s′.

          * Actualizar Q(s,a) usando la fórmula de actualización.

          * Establecer s←s′.

3. Extraer la política:

      * Una vez entrenado, la política óptima se deriva seleccionando la acción con mayor valor en cada estado:
<center>π(s)=arg max<sub>a</sub>Q(s,a)</center>


**Características clave**

**Off-policy**: Aprende sobre la política óptima aunque actúe siguiendo una política distinta (por ejemplo, exploratoria).

**Convergencia**: Garantizada bajo ciertas condiciones (tasa de aprendizaje adecuada, exploración suficiente).

**Limitaciones**: No escala bien a espacios grandes o continuos, donde se requieren funciones de aproximación (como redes neuronales en el Deep Q-Learning).

####Implementación de Q-Learning desde 0

Voy a utilizar la biblioteca Gymnasium, que acabamos de conocer, y tras consultar distintas fuentes, creo que lo mejor será usar el entorno FrozenLake-v1.

````python
import numpy as np
import gymnasium as gym
import random

# Crear entorno determinista
env = gym.make("FrozenLake-v1", is_slippery=False)
env = gym.wrappers.TimeLimit(env, max_episode_steps=100)  # Asegura episodios largos
print("Configuración del entorno:", env.spec.kwargs)

# Parámetros del entorno
n_states = env.observation_space.n
n_actions = env.action_space.n

# Inicialización de la tabla Q
Q = np.zeros((n_states, n_actions))

# Hiperparámetros
alpha = 0.1
gamma = 0.99
epsilon = 1.0
epsilon_min = 0.1
epsilon_decay = 0.999
episodes = 10000

# Entrenamiento Q-Learning
for episode in range(episodes):
    state, _ = env.reset()
    done = False

    while not done:
        # Política epsilon-greedy
        if random.uniform(0, 1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(Q[state, :])

        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        # Actualización Q
        Q[state, action] += alpha * (
            reward + gamma * np.max(Q[next_state, :]) - Q[state, action]
        )

        state = next_state

    # Decaimiento de epsilon
    epsilon = max(epsilon_min, epsilon * epsilon_decay)
````

####Relación con el anterior apartado

La implementación sigue paso a paso la descripción teórica del algoritmo:

1. Se inicializa una tabla Q(s,a) con ceros.

2. Para cada episodio, el agente comienza en un estado inicial y actúa siguiendo una política ε-greedy, tal como se describió anteriormente.

3. La fórmula de actualización:
<center>Q(s,a)←Q(s,a)+α[r+γmax<sub>a′</sub>Q(s′,a′)−Q(s,a)]</center>
se aplica en cada transición, ajustando el valor estimado de la acción tomada.

El agente aprende de manera diferente a la que actua (*off-policy*), ya que aunque elige acciones de forma exploratoria, actualiza el valor Q asumiendo que tomará la mejor acción posible en el futuro.

En el Aprendizaje por refuerzo, el objetivo del agente es maximizar la recompensa acumulada al interactuar con un entorno. Para conseguirlo, el agente sigue una política, que es la estrategia que define qué acción tomar en cada estado. Los dos quieren optimizar el comportamiento de un agente en un entorno para maximizar la recompensa pero lo hacen de forma diferente. De hecho, existen dos grandes enfoques para aprender esta política de manera óptima:

* Métodos basados en valor: estiman el valor de los estados o de los pares estado-acción. Derivan la política indirectamente seleccionando las acciones con el valor más alto, es decir, no aprenden la política directamente. La política que conduce a la función de valor óptima se elige como política óptima. Las ecuaciones de Bellman describen las funciones estado-valor y las funciones estado-acción-valor óptimas. Hablaremos de **Deep Q-Networks (DQN)**.

* Métodos basados en políticas: aprenden directamente la política óptima aplicando el ascenso de gradiente sobre la recompensa esperada. No se basan en una función de valor. La política se expresa en forma parametrizada. Cuando la política se aplica mediante una red neuronal, los parámetros de la política se refieren a los pesos de la red. La red aprende la política óptima utilizando el ascenso gradiente sobre los parámetros de la política. Hablaremos de **Policy Gradient (PG)**.



Primero vamos a hablar de DQN. En el año 2015, DeepMind presentó DQN y así nació el Deep Reinforcement Learning. Originalmente se aplicó a videojuegos de Atari, donde obtuvo resultados superiores a los de humanos expertos.

Este algoritmo combina el algoritmo Q-learning con redes neuronales profundas. Sabemos que las redes neuronales se utilizan para aproximar funciones no lineales, entonces con este algoritmo usamos una red neuronal para aproximar la función Q, así evitamos utilizar una tabla. Es decir, nos enfocamos en aprender una función de valor, denominada Q-value. Este valor representa la recompensa acumulada al tomar una acción $a$ en un estado $s$ y luego seguir una política óptima.

En el aprendizaje por refuerzo clásico, el agente tiene una función de valor llamada **Q-function**, que estima el valor de tomar una acción \(a\) en un estado \(s\) siguiendo una política óptima:

$$
Q^*(s, a) = \mathbb{E} \left[ r + \gamma \max_{a'} Q^*(s', a') \mid s, a \right]
$$

Donde:  
- r es la recompensa recibida al ejecutar la acción a en el estado s.  
- $\gamma$ es el factor de descuento ($0 < \gamma \leq 1$), que pondera las recompensas futuras.  
- s' es el siguiente estado al tomar la acción a.  
- a' son las posibles acciones en el siguiente estado.

Para cada transición, se calcula el valor objetivo:

\begin{equation}
    y = r + \gamma \max_{a'} Q'(s', a')
\end{equation}

Donde:
* $Q'$ es la predicción de la Target Network (estabiliza el entrenamiento y evita oscilaciones durante el aprendizaje).
* $y$ es el valor que debería tener $Q(s, a)$ para minimizar el error.


En lugar de almacenar una tabla Q(s, a), **DQN usa una red neuronal** que toma como entrada un estado s y predice el valor Q(s, a) para cada acción posible.  
- La entrada de la red es el estado s, que puede ser una imagen (en juegos Atari, por ejemplo) o un vector de características.  
- La salida de la red son los valores Q(s, a) para cada acción.  
- La red se entrena para minimizar el error entre el valor predicho y el valor objetivo, definido por la **Ecuación de Bellman**.  


Por otro lado tenemos Policy Gradient (PG) que optimiza directamente la política de un agente en lugar de aprender una función de valor como hace DQN. Una política es simplemente una función que define el comportamiento del agente, es decir, la probabilidad de elegir una acción $a$ dado un estado
$s$:

$$
\pi(a \mid s, \theta) = P(a \mid s)
$$

Donde:
* $\pi$ es la política del agente.
* $s$ es el estado actual.
* $a$ es la acción seleccionada.
* $\theta$ son los parámetros de la red neuronal que representan esa política.

Una de las razones por las que se utiliza esta categoría es porque muchas veces es difícil y costoso estimar la función de valor para todas las combinaciones de estados y acciones. En lugar de eso, PG aprende directamente la probabilidad de tomar cada acción en un estado dado, optimizando los parámetros $\theta$ de una red neuronal para maximizar la recompensa acumulada.

Policy Gradient optimiza esta política directamente aplicando el Gradiente del Rendimiento Esperado sobre los parámetros $\theta$ de la red neuronal:

$$
\nabla_\theta J(\theta) = \mathbb{E} \left[\nabla_\theta \log \pi_{\theta}(a | s) \cdot R \right]
$$

Donde:

* $J(\theta)$ es la función objetivo (la recompensa esperada acumulada).
* $R$ es el retorno total obtenido en esa trayectoria.

**Generación del conjunto de datos (Replay Buffer):**

En DQN, el conjunto de datos de entrada y salida se genera de manera dinámica mediante la interacción del agente con el entorno.

* Cada transición $(s, a, r, s')$ se almacena en el Replay Buffer.
* Durante el entrenamiento, se extrae un minibatch aleatorio de este buffer para entrenar la red neuronal.
* Este muestreo aleatorio rompe la correlación temporal entre transiciones consecutivas, mejorando la estabilidad del entrenamiento y evitando sesgos.

**Entrenamiento de la Red Neuronal:**

Cada cierto número de pasos, se entrena la red neuronal con un minibatch de transiciones seleccionadas aleatoriamente del Replay Buffer.  
Para cada transición $(s, a, r, s')$, se calcula el valor objetivo (target value) según la ecuación de Bellman:

$$
y =
\begin{cases}
    r & \text{si el episodio termina en } s' \\
    r + \gamma \max_{a'} Q_{\text{target}}(s', a') & \text{en caso contrario}
\end{cases}
$$

donde:
* $y$ es el valor objetivo que debería tener $Q(s, a)$ para minimizar el error.
* $\gamma$ es el factor de descuento.
* $Q_{\text{target}}$ es el valor predicho por la Target Network, que se actualiza periódicamente para evitar oscilaciones bruscas.

**Función de pérdida:**

La red neuronal se entrena minimizando el error entre el valor predicho $Q(s, a)$ y el valor objetivo $y$. Para ello, se utiliza la función de pérdida cuadrática media (MSE):

$$
L(\theta) = \mathbb{E} \left[ \left(y - Q(s, a; \theta) \right)^2 \right]
$$

Donde:
* $\theta$ son los parámetros (pesos) de la red neuronal.
* $y$ es el valor objetivo calculado en el paso anterior.
* $Q(s, a; \theta)$ es la predicción de la red para el par $(s, a)$.


Esta pérdida se optimiza usando técnicas clásicas de optimización como Stochastic Gradient Descent (SGD) o Adam Optimizer.

**Actualización de la Target Network:**

Cada $N$ pasos, los pesos de la red principal (Q-Network) se copian en la Target Network:

$$
\theta_{\text{target}} \leftarrow \theta
$$

Esto estabiliza el entrenamiento, ya que la red que evalúa los valores futuros no está cambiando constantemente, lo cual reduce las oscilaciones y facilita la convergencia.

Importamos nuestro juego del taxi:

````python
import tensorflow as tf
from tf_agents.environments import suite_gym
from tf_agents.environments import tf_py_environment
from tf_agents.networks import q_network
from tf_agents.agents.dqn import dqn_agent
from tf_agents.replay_buffers import tf_uniform_replay_buffer
from tf_agents.trajectories import trajectory
from tf_agents.utils import common
import gymnasium as gym
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
````

````python
#Importamos el entorno del taxi
gym.make("Taxi-v3")
````

````python
#Creamos el entorno
env = suite_gym.load('Taxi-v3')
train_env = tf_py_environment.TFPyEnvironment(env)
eval_env = tf_py_environment.TFPyEnvironment(env)
````

````python
#Definimos la red neuronal (Q-Network)
q_net = q_network.QNetwork(
    train_env.observation_spec(),
    train_env.action_spec(),
    fc_layer_params=(64, 64)
)
````

````python
#Configuramos el agente DQN
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)

agent = dqn_agent.DqnAgent(
    train_env.time_step_spec(),
    train_env.action_spec(),
    q_network=q_net,
    optimizer=optimizer,
    td_errors_loss_fn=common.element_wise_squared_loss,
    train_step_counter=tf.Variable(0)
)

agent.initialize() #lo inicializamos
````

````python
#Definimos el buffer de memoria (replay buffer)
replay_buffer = tf_uniform_replay_buffer.TFUniformReplayBuffer(
    data_spec=agent.collect_data_spec,
    batch_size=train_env.batch_size,
    max_length=10000
)
````

````python
#Definimos la funcion para almacenar las transiciones
def collect_step(environment, policy, buffer):
    time_step = environment.current_time_step()
    action_step = policy.action(time_step)
    next_time_step = environment.step(action_step.action)
    traj = trajectory.from_transition(time_step, action_step, next_time_step)
    buffer.add_batch(traj)
````

````python
# Funcion para entrenar
def train_agent():
    dataset = replay_buffer.as_dataset(
        num_parallel_calls=3,
        sample_batch_size=64,
        num_steps=2
    ).prefetch(3)

    iterator = iter(dataset)
    for _ in range(10):
        experience, _ = next(iterator)
        agent.train(experience)
````

````python
#datos del entrenamiento
num_episodes = 100
rewards = []
````

````python
# Entrenamiento
for episode in tqdm(range(num_episodes)):
    time_step = train_env.reset()
    episode_reward = 0

    while not time_step.is_last():
        collect_step(train_env, agent.collect_policy, replay_buffer)  # POLÍTICA DE EXPLORACIÓN
        time_step = train_env.current_time_step()
        episode_reward += time_step.reward.numpy()[0]

        # Entrenar el agente
        if replay_buffer.num_frames() > 64:
            train_agent()

    print(f"Episode {episode + 1}: Recompensa -> {episode_reward}")
    rewards.append(episode_reward)
````

````python
#Mostramos los resultados
df_rewards = pd.DataFrame(rewards, columns=["Recompensa"])
````
